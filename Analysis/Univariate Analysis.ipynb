```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import datetime as dt
```

```python
cn=pd.read_csv(r"..\Data Sets\cleaned_novel_dataset.csv")
```

```python
cn["Publish_Date"]=pd.to_datetime(cn["Publish_Date"])
```

```python
#Dataset without duplicate values with respect to Novel_Name, Series_Name, Author_Name
without_duplicates=cn.drop_duplicates(subset=['Novel_Name', 'Series_Name', 'Author_Name'],keep="first")
without_duplicates
```

```python
# It gives the count of the novel of each genre in the dataset
pd.DataFrame(cn["Genre"].value_counts())
```

```python
# Creating barplot for the count of each Genre
plt.figure(figsize=(8,4))
ax=sns.countplot(x="Genre",data=cn,palette=sns.color_palette("Spectral",9))
plt.title("Count of each Genre")
plt.ylim([0,900])
plt.xticks(rotation=90)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',(p.get_x()+p.get_width()/2,p.get_height()+15),ha="center")
plt.show()
```

From the above bar plot, we can analyze the distribution of the genre in our dataset. It can be seen that all the data are almost evenly distributed between 787 and 800, apart from "Historical" and "Suspence/Thriller" which has 735 and 693 datas resp.

```python
#It gives the top 20 years in which there are more number of novels published in the dataset
pd.DataFrame(without_duplicates["Publish_year"].value_counts().head(20))
```

```python
# Plotting histogram of Publish Year
plt.figure(figsize=(8,4))
ax=sns.histplot(x=without_duplicates["Publish_year"],binwidth=10,stat="count",kde=True,palette="flare")
plt.title("Histogram of Publish Year")
plt.ylim([0,2500])
plt.xticks(rotation=90)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',(p.get_x()+p.get_width()/2,p.get_height()+15),ha="center")
plt.show()
```

```python
#Plotting KDE of Publish Year with mean, median and mode
plt.figure(figsize=(8,4))
sns.kdeplot(without_duplicates["Publish_year"])
plt.title("Kernel Density Estimate of Publish Year")
plt.axvline(without_duplicates["Publish_year"].mean(),color="green",label="Mean")
plt.axvline(without_duplicates["Publish_year"].median(),color="red",label="Median")
plt.axvline(without_duplicates["Publish_year"].mode()[0],color="purple",label="Mode")
plt.legend()
plt.grid()
plt.show()
```

From the above kde plot, it can be observed that most of the novels in the dataset are published after 1980.
