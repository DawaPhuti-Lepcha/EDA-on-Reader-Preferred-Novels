### Dropping and replacing

```python
#Replacing the null values of "Series_Name" column with '-' since it is an object type.
novel["Series_Name"].fillna('-',inplace=True)
```

```python
# dropping the "Pages" column since there are 149 missing values and a lot of data are incorrecct.
novel.drop(["Pages"],axis=1,inplace=True)
```

```python
novel.isna().sum()
```

```python
#Checking the number and percentage of null values 
print(novel.isna().sum().sum())
print(novel.isna().sum().sum()/novel.shape[0]*100)
```

```python
#Dropping the null values since there are less than 1% null values in the dataset
novel.dropna(axis=0,inplace=True)
novel=novel.reset_index(drop=True)
novel
```

### Removing the comma and  changing the datatypes.

```python
novel["No_of_Ratings"]=novel["No_of_Ratings"].str.replace(',','').astype('int')
```

```python
novel["No_of_Reviews"]=novel["No_of_Reviews"].str.replace(',','').astype('int')
```

```python
novel["No_of_Votes"]=novel["No_of_Votes"].str.replace(',','').astype('int')
```

```python
novel["Score"]=novel["Score"].str.replace(',','').astype('int')
```

### Formatting the 'Publish_Year' and changing its datatype

```python
import calendar
month_dict = dict((v, k) for k, v in enumerate(calendar.month_name))
```

```python
month_change=lambda x: month_dict[x]
```

```python
novel[["month","date","year"]]=novel["Publish_Year"].str.split(' ',expand=True)
```

```python
novel["date"]=novel["date"].str.replace(',','')
```

```python
novel["month"]=novel["month"].apply(month_change).astype("str")
```

```python
novel["Publish_Year"]=novel["date"]+'-'+novel["month"]+'-'+novel["year"]
```

```python
novel.drop(novel.loc[:,'month':'year'],axis=1,inplace=True)
```

```python
#changes the datatype to datetime and gives null values to the datas which are not convertible 
novel["Publish_Year"]=pd.to_datetime(novel["Publish_Year"],errors='coerce')
novel
```

```python
novel.info()
```
